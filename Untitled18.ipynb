```python
import os

# Find all Python files in realcause that use DataFrame.append
realcause_root = "/content/realcause"  # adjust if cloned elsewhere

patch_count = 0
for dirpath, _, filenames in os.walk(realcause_root):
    for fname in filenames:
        if not fname.endswith(".py"):
            continue
        fpath = os.path.join(dirpath, fname)
        with open(fpath, "r") as f:
            src = f.read()
        if ".append(" not in src:
            continue
        # Replace df.append(other, ignore_index=True) → pd.concat([df, other], ignore_index=True)
        # This covers the most common patterns in RealCause
        import re
        patched = re.sub(
            r'(\w+)\.append\((\w+),\s*ignore_index=True\)',
            r'pd.concat([\1, \2], ignore_index=True)',
            src
        )
        patched = re.sub(
            r'(\w+)\.append\((\w+)\)',
            r'pd.concat([\1, \2], ignore_index=True)',
            patched
        )
        if patched != src:
            with open(fpath, "w") as f:
                f.write(patched)
            print(f"Patched: {fpath}")
            patch_count += 1

print(f"\nDone — {patch_count} files patched")
```

```python
import re
from pathlib import Path

def patch_file(fpath: str):
    path = Path(fpath)
    if not path.exists():
        print(f"NOT FOUND: {fpath}")
        return

    src = path.read_text()
    original = src

    # Pattern 1: df = df.append(row, ignore_index=True)
    src = re.sub(
        r'(\w+)\s*=\s*(\w+)\.append\((.+?),\s*ignore_index\s*=\s*True\)',
        r'\1 = pd.concat([\2, pd.DataFrame([\3])], ignore_index=True)',
        src
    )

    # Pattern 2: df.append(other_df, ignore_index=True)
    src = re.sub(
        r'(\w+)\.append\((\w+),\s*ignore_index\s*=\s*True\)',
        r'pd.concat([\1, \2], ignore_index=True)',
        src
    )

    # Pattern 3: df.append(other_df) — no ignore_index
    src = re.sub(
        r'(\w+)\.append\((\w+)\)',
        r'pd.concat([\1, \2], ignore_index=True)',
        src
    )

    # Pattern 4: .append({dict}) — appending a dict row
    src = re.sub(
        r'(\w+)\.append\((\{.+?\}),\s*ignore_index\s*=\s*True\)',
        r'pd.concat([\1, pd.DataFrame([\2])], ignore_index=True)',
        src
    )

    # Ensure pd is imported if we made changes
    if src != original and "import pandas as pd" not in src:
        src = "import pandas as pd\n" + src

    if src != original:
        path.write_text(src)
        print(f"✓ Patched: {fpath}")
    else:
        print(f"  No changes: {fpath}")

# Patch all known RealCause files that use append
import os
realcause_root = "/content/realcause"
for dirpath, _, filenames in os.walk(realcause_root):
    for fname in filenames:
        if fname.endswith(".py"):
            patch_file(os.path.join(dirpath, fname))
```

```python
import os
import sys
import pandas as pd
import numpy as np

os.chdir("/content/realcause")
sys.path.insert(0, "/content/realcause")

# ── Correct filename pattern ──────────────────────────────────────────────────
# Files are: lalonde_psid_sample6.csv  (NOT lalonde_psid_sample_6.csv)

def load_all_samples(dataset: str, n_samples: int = 100) -> list[pd.DataFrame]:
    samples = []
    missing = []
    for i in range(n_samples):
        # No underscore between 'sample' and the number
        path = f"/content/realcause/realcause_datasets/{dataset}_sample{i}.csv"
        if os.path.exists(path):
            df = pd.read_csv(path)
            df["sample_id"] = i
            df["dataset"]   = dataset
            samples.append(df)
        else:
            missing.append(i)
    if missing:
        print(f"  Missing samples: {missing}")
    print(f"  Loaded {len(samples)}/100 samples for '{dataset}'")
    return samples

psid_samples = load_all_samples("lalonde_psid")
cps_samples  = load_all_samples("lalonde_cps")
twin_samples = load_all_samples("twins")
```

```python
# ── Now run your causal estimators on these samples ──────────────────────────
# This is the actual benchmarking pipeline

from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.ensemble import RandomForestRegressor
import warnings
warnings.filterwarnings("ignore")

COVARIATE_COLS = {
    "lalonde_psid": ["age", "education", "black", "hispanic",
                     "married", "nodegree", "re74", "re75"],
    "lalonde_cps":  ["age", "education", "black", "hispanic",
                     "married", "nodegree", "re74", "re75"],
    "twins":        [c for c in twin_samples[0].columns
                     if c not in ["t", "y", "y0", "y1", "ite",
                                  "sample_id", "dataset"]],
}

def naive_ate(df, covs):
    """Simple difference in means — no adjustment."""
    return df.loc[df["t"]==1, "y"].mean() - df.loc[df["t"]==0, "y"].mean()

def ols_ate(df, covs):
    """OLS with covariate adjustment (your ANCOVA estimator)."""
    X = pd.get_dummies(df[covs + ["t"]], drop_first=True).astype(float)
    y = df["y"].values
    model = LinearRegression().fit(X, y)
    return model.coef_[X.columns.get_loc("t")]

def ipw_ate(df, covs):
    """Inverse probability weighting."""
    X = pd.get_dummies(df[covs], drop_first=True).astype(float)
    t = df["t"].values
    y = df["y"].values
    ps = LogisticRegression(max_iter=1000).fit(X, t).predict_proba(X)[:, 1]
    ps = np.clip(ps, 0.01, 0.99)
    ipw = np.where(t==1, y/ps, -y/(1-ps))
    return ipw.mean()

def aipw_ate(df, covs):
    """Augmented IPW (doubly robust)."""
    X  = pd.get_dummies(df[covs], drop_first=True).astype(float)
    t  = df["t"].values
    y  = df["y"].values
    ps = LogisticRegression(max_iter=1000).fit(X, t).predict_proba(X)[:, 1]
    ps = np.clip(ps, 0.01, 0.99)
    X1 = X.copy(); X1["t"] = 1
    X0 = X.copy(); X0["t"] = 0
    Xt = pd.get_dummies(df[covs + ["t"]], drop_first=True).astype(float)
    or_model = LinearRegression().fit(Xt, y)
    mu1 = or_model.predict(X1)
    mu0 = or_model.predict(X0)
    dr = (mu1 - mu0
          + t*(y - mu1)/ps
          - (1-t)*(y - mu0)/(1-ps))
    return dr.mean()

ESTIMATORS = {
    "Naive (diff-in-means)": naive_ate,
    "OLS/ANCOVA":            ols_ate,
    "IPW":                   ipw_ate,
    "AIPW (doubly robust)":  aipw_ate,
}

print("Estimators registered:", list(ESTIMATORS.keys()))
```

```python
# ── Run benchmark across all samples and datasets ────────────────────────────

results = []

for dataset, samples in [("lalonde_psid", psid_samples),
                          ("lalonde_cps",  cps_samples),
                          ("twins",        twin_samples)]:
    covs = COVARIATE_COLS[dataset]
    print(f"\nBenchmarking {dataset} ({len(samples)} samples)...")

    for df in samples:
        true_ate = df["ite"].mean()

        for est_name, est_fn in ESTIMATORS.items():
            try:
                est_ate = est_fn(df, covs)
                results.append({
                    "dataset":    dataset,
                    "sample_id":  df["sample_id"].iloc[0],
                    "estimator":  est_name,
                    "true_ate":   true_ate,
                    "est_ate":    est_ate,
                    "bias":       est_ate - true_ate,
                    "abs_bias":   abs(est_ate - true_ate),
                    "sq_error":   (est_ate - true_ate) ** 2,
                })
            except Exception as e:
                print(f"  ✗ {est_name} on sample {df['sample_id'].iloc[0]}: {e}")

results_df = pd.DataFrame(results)
print(f"\nTotal results: {len(results_df):,} rows")
results_df.head(10)
```

```python
# ── Summary table: ATE bias and RMSE per estimator per dataset ───────────────

summary = (
    results_df
    .groupby(["dataset", "estimator"])
    .agg(
        mean_bias   = ("bias",     "mean"),
        abs_bias    = ("abs_bias", "mean"),
        rmse        = ("sq_error", lambda x: np.sqrt(x.mean())),
        n_samples   = ("sample_id","count"),
    )
    .round(2)
    .reset_index()
)

print("=== Benchmark Summary ===")
print(summary.to_string(index=False))

# Save for paper
summary.to_csv("benchmark_results.csv", index=False)
results_df.to_csv("benchmark_results_full.csv", index=False)
print("\nSaved → benchmark_results.csv")
```

```python
from econml.dml import CausalForestDML
from econml.metalearners import XLearner, TLearner
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier, GradientBoostingRegressor
from sklearn.linear_model import LogisticRegression, LinearRegression
import warnings
warnings.filterwarnings("ignore")

def causal_forest_ate(df, covs):
    X = pd.get_dummies(df[covs], drop_first=True).astype(float).values
    t = df["t"].values.astype(float)
    y = df["y"].values.astype(float)
    cf = CausalForestDML(
        model_y           = RandomForestRegressor(n_estimators=100, random_state=42),
        model_t           = RandomForestClassifier(n_estimators=100, random_state=42),
        discrete_treatment = True,   # ← key fix: tells DML that t is binary
        n_estimators      = 100,
        random_state      = 42,
    )
    cf.fit(y, t, X=X)
    return float(cf.ate(X))


def xlearner_ate(df, covs):
    X = pd.get_dummies(df[covs], drop_first=True).astype(float).values
    t = df["t"].values.astype(float)
    y = df["y"].values.astype(float)
    xl = XLearner(
        models           = GradientBoostingRegressor(random_state=42),
        propensity_model = LogisticRegression(max_iter=500),
    )
    xl.fit(y, t, X=X)
    # ← key fix: .ate() takes only X in newer econml, no T keyword
    return float(xl.ate(X))


def tlearner_ate(df, covs):
    X = pd.get_dummies(df[covs], drop_first=True).astype(float).values
    t = df["t"].values.astype(float)
    y = df["y"].values.astype(float)
    tl = TLearner(models=GradientBoostingRegressor(random_state=42))
    tl.fit(y, t, X=X)
    return float(tl.ate(X))


# Rebuild the full estimator registry
ESTIMATORS_EXTENDED = {
    "Naive (diff-in-means)": naive_ate,
    "IPW":                   ipw_ate,
    "OLS/ANCOVA":            ols_ate,
    "AIPW (doubly robust)":  aipw_ate,
    "T-Learner":             tlearner_ate,
    "X-Learner":             xlearner_ate,
    "Causal Forest":         causal_forest_ate,
}

print("Estimators ready:", list(ESTIMATORS_EXTENDED.keys()))
```

```python
# ── Run extended benchmark (use subset for speed — 20 samples per dataset) ───

results_ext = []
N_SAMPLES   = 100   # increase to 100 for final paper run

for dataset, samples in [("lalonde_psid", psid_samples[:N_SAMPLES]),
                          ("lalonde_cps",  cps_samples[:N_SAMPLES]),
                          ("twins",        twin_samples[:N_SAMPLES])]:
    covs = COVARIATE_COLS[dataset]
    print(f"\nBenchmarking {dataset} ({N_SAMPLES} samples)...")

    for s_idx, df in enumerate(samples):
        true_ate = df["ite"].mean()

        for est_name, est_fn in ESTIMATORS_EXTENDED.items():
            try:
                est_ate = float(est_fn(df, covs))
                results_ext.append({
                    "dataset":   dataset,
                    "sample_id": df["sample_id"].iloc[0],
                    "estimator": est_name,
                    "true_ate":  true_ate,
                    "est_ate":   est_ate,
                    "bias":      est_ate - true_ate,
                    "abs_bias":  abs(est_ate - true_ate),
                    "sq_error":  (est_ate - true_ate) ** 2,
                })
            except Exception as e:
                print(f"  ✗ {est_name} sample {s_idx}: {e}")

        if (s_idx + 1) % 5 == 0:
            print(f"  {s_idx+1}/{N_SAMPLES} done")

results_ext_df = pd.DataFrame(results_ext)
print(f"\nTotal: {len(results_ext_df):,} rows")
```

```python
# ── Summary ───────────────────────────────────────────────────────────────────

summary_ext = (
    results_ext_df
    .groupby(["dataset", "estimator"])
    .agg(
        mean_bias = ("bias",     "mean"),
        abs_bias  = ("abs_bias", "mean"),
        rmse      = ("sq_error", lambda x: np.sqrt(x.mean())),
        n         = ("sample_id","count"),
    )
    .round(2)
    .reset_index()
)
print(summary_ext.to_string(index=False))
```

```python
# ── Figure 5: Variance vs Bias decomposition ──────────────────────────────────
# Shows that high RMSE of ML estimators is driven by variance, not bias

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for ax, (dataset, label) in zip(axes, [
    ("lalonde_psid", "LaLonde PSID"),
    ("lalonde_cps",  "LaLonde CPS"),
]):
    sub = summary_ext[summary_ext["dataset"] == dataset].copy()
    sub["variance_component"] = sub["rmse"]**2 - sub["mean_bias"]**2
    sub["variance_component"] = sub["variance_component"].clip(lower=0)
    sub["std_component"] = np.sqrt(sub["variance_component"])
    sub["complexity"] = sub["estimator"].map(COMPLEXITY)
    sub = sub.sort_values("complexity")

    x      = np.arange(len(sub))
    width  = 0.38

    bars1 = ax.bar(x - width/2, sub["abs_bias"].abs(), width,
                   label="|Bias|", color="#E07B54", alpha=0.85, edgecolor="white")
    bars2 = ax.bar(x + width/2, sub["std_component"], width,
                   label="Std Dev", color="#5B8DB8", alpha=0.85, edgecolor="white")

    ax.set_xticks(x)
    ax.set_xticklabels(
        [COMPLEXITY_LABELS[sub["complexity"].iloc[i]] for i in range(len(sub))],
        fontsize=9
    )
    ax.set_ylabel("Dollars ($)", fontsize=11)
    ax.set_title(label, fontweight="bold", fontsize=13)
    ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"${x:,.0f}"))
    ax.legend(fontsize=10, framealpha=0)

fig.suptitle(
    "Bias–Variance Decomposition of ATE Estimation Error\n"
    "|Bias| vs Standard Deviation Across 100 Samples",
    fontweight="bold", fontsize=13, y=1.03
)
plt.tight_layout()
plt.savefig("fig5_bias_variance.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved → fig5_bias_variance.png")
```

```python
# ── Wilcoxon signed-rank tests: pairwise comparisons ─────────────────────────
from scipy.stats import wilcoxon
import pandas as pd
import numpy as np

# Run on BOTH LaLonde datasets
for dataset in ["lalonde_psid", "lalonde_cps"]:
    print(f"\n{'='*70}")
    print(f"Wilcoxon signed-rank tests — {dataset.upper()}")
    print(f"H₀: no difference in |bias| between T-Learner and competitor")
    print(f"{'='*70}")
    print(f"{'Comparison':<35} {'T-Lrn |Bias|':>13} {'Comp |Bias|':>12} {'p-value':>10} {'Sig':>5} {'Direction'}")
    print("-"*70)

    # Extract T-Learner absolute bias for this dataset
    tlearner_bias = results_ext_df[
        (results_ext_df["dataset"]   == dataset) &
        (results_ext_df["estimator"] == "T-Learner")
    ]["abs_bias"].values

    competitors = [
        "Naive (diff-in-means)",
        "IPW",
        "OLS/ANCOVA",
        "AIPW (doubly robust)",
        "X-Learner",
        "Causal Forest",
    ]

    n_comparisons = len(competitors)
    raw_pvals = []
    results_rows = []

    for comp in competitors:
        comp_bias = results_ext_df[
            (results_ext_df["dataset"]   == dataset) &
            (results_ext_df["estimator"] == comp)
        ]["abs_bias"].values

        # Align lengths (should both be 100)
        min_len = min(len(tlearner_bias), len(comp_bias))
        t_vals  = tlearner_bias[:min_len]
        c_vals  = comp_bias[:min_len]

        try:
            stat, p_raw = wilcoxon(t_vals, c_vals, alternative="two-sided")
            raw_pvals.append(p_raw)
            results_rows.append((comp, t_vals.mean(), c_vals.mean(), stat, p_raw))
        except Exception as e:
            print(f"  ERROR on {comp}: {e}")
            raw_pvals.append(1.0)
            results_rows.append((comp, t_vals.mean(), c_vals.mean(), None, 1.0))

    # Bonferroni correction
    bonferroni_pvals = [min(p * n_comparisons, 1.0) for p in raw_pvals]

    for (comp, t_mean, c_mean, stat, p_raw), p_bonf in zip(results_rows, bonferroni_pvals):
        sig = "***" if p_bonf < 0.001 else "**" if p_bonf < 0.01 else "*" if p_bonf < 0.05 else "ns"
        direction = "T-Learner BETTER" if t_mean < c_mean else "T-Learner WORSE"
        print(f"  T-Learner vs {comp:<22} {t_mean:>12,.1f} {c_mean:>12,.1f} "
              f"{p_bonf:>10.4f} {sig:>5}  {direction}")
```

```python
# ── Full summary table for paper ─────────────────────────────────────────────

print("\n\n=== PAPER-READY TABLE (LaLonde CPS) ===")
print(f"{'Comparison':<35} {'T-Lrn |Bias|':>14} {'Comp |Bias|':>13} "
      f"{'p (raw)':>10} {'p (Bonf.)':>10} {'Sig':>5}")
print("-"*90)

dataset = "lalonde_cps"
tlearner_bias = results_ext_df[
    (results_ext_df["dataset"]   == dataset) &
    (results_ext_df["estimator"] == "T-Learner")
]["abs_bias"].values

competitors = [
    "Naive (diff-in-means)", "IPW", "OLS/ANCOVA",
    "AIPW (doubly robust)", "X-Learner", "Causal Forest"
]

raw_pvals, rows = [], []
for comp in competitors:
    comp_bias = results_ext_df[
        (results_ext_df["dataset"]   == dataset) &
        (results_ext_df["estimator"] == comp)
    ]["abs_bias"].values
    min_len = min(len(tlearner_bias), len(comp_bias))
    try:
        stat, p = wilcoxon(tlearner_bias[:min_len], comp_bias[:min_len])
        raw_pvals.append(p)
        rows.append((comp, tlearner_bias[:min_len].mean(), comp_bias[:min_len].mean(), p))
    except Exception as e:
        raw_pvals.append(1.0)
        rows.append((comp, 0, 0, 1.0))

bonf = [min(p * len(competitors), 1.0) for p in raw_pvals]

for (comp, tm, cm, pr), pb in zip(rows, bonf):
    sig = "***" if pb < 0.001 else "**" if pb < 0.01 else "*" if pb < 0.05 else "ns"
    pr_str = f"<0.001" if pr < 0.001 else f"{pr:.4f}"
    pb_str = f"<0.001" if pb < 0.001 else f"{pb:.4f}"
    print(f"  T-Learner vs {comp:<22} ${tm:>12,.0f} ${cm:>12,.0f} "
          f"{pr_str:>10} {pb_str:>10} {sig:>5}")
```

```python
# ── Additional test: is T-Learner significantly better than Naive on PSID? ───
# (Important — Naive is the overall winner on PSID so this tells us
#  whether T-Learner's gap from Naive is statistically meaningful)

print("\n=== T-Learner vs Naive on LaLonde PSID ===")
for dataset in ["lalonde_psid", "lalonde_cps"]:
    tl = results_ext_df[
        (results_ext_df["dataset"]   == dataset) &
        (results_ext_df["estimator"] == "T-Learner")
    ]["abs_bias"].values

    naive = results_ext_df[
        (results_ext_df["dataset"]   == dataset) &
        (results_ext_df["estimator"] == "Naive (diff-in-means)")
    ]["abs_bias"].values

    min_len = min(len(tl), len(naive))
    stat, p = wilcoxon(tl[:min_len], naive[:min_len])
    sig = "***" if p < 0.001 else "**" if p < 0.01 else "*" if p < 0.05 else "ns"
    winner = "Naive BETTER" if naive.mean() < tl.mean() else "T-Learner BETTER"
    print(f"  {dataset:<15}  Naive |bias|=${naive.mean():,.0f}  "
          f"T-Learner |bias|=${tl.mean():,.0f}  "
          f"p={p:.4f} {sig}  → {winner}")
```

```python
# ── Effect size (rank-biserial correlation) for each comparison ───────────────
# r = 1 - 2U/(n1*n2) — gives a standardised effect size alongside p-value

from scipy.stats import mannwhitneyu

print("\n=== Effect sizes (rank-biserial r) — LaLonde CPS ===")
print("r > 0.5 = large effect, 0.3-0.5 = medium, 0.1-0.3 = small")
print("-"*65)

dataset = "lalonde_cps"
tlearner_bias = results_ext_df[
    (results_ext_df["dataset"]   == dataset) &
    (results_ext_df["estimator"] == "T-Learner")
]["abs_bias"].values

for comp in competitors:
    comp_bias = results_ext_df[
        (results_ext_df["dataset"]   == dataset) &
        (results_ext_df["estimator"] == comp)
    ]["abs_bias"].values
    min_len = min(len(tlearner_bias), len(comp_bias))
    U, _ = mannwhitneyu(tlearner_bias[:min_len], comp_bias[:min_len],
                        alternative="two-sided")
    n = min_len
    r = 1 - (2 * U) / (n * n)
    magnitude = "large" if abs(r) > 0.5 else "medium" if abs(r) > 0.3 else "small"
    print(f"  T-Learner vs {comp:<25}  r = {r:+.3f}  ({magnitude})")
```
